```python
import io
import sqlite3
import pandas as pd
```

```python
conn = sqlite3.connect('copas.db')
cursor = conn.cursor()
```

```python
f = io.open('database.sql', 'r', encoding='ISO-8859-1')
sql = f.read()
cursor.executescript(sql)
```

```python
query = """
SELECT DISTINCT nome_jogador
FROM jogadores
WHERE numero_camisa = 14
"""

df = pd.read_sql_query(query, conn)
df
```

```python
query = """
SELECT DISTINCT estadio
FROM local natural join partidas
WHERE partidas.ano = 1958
"""

df = pd.read_sql_query(query, conn)
df
```

```python
query = """
SELECT DISTINCT nome_treinador
FROM partidas natural join jogadores
WHERE partidas.gols_mandante >= 7
"""

df = pd.read_sql_query(query, conn)
df
```

```python
query = """
SELECT DISTINCT nome_jogador
FROM jogadores natural join partidas
WHERE iniciais_time = 'ITA' AND ano = 1994
"""

df = pd.read_sql_query(query, conn)
df
```

```python
query = """
SELECT DISTINCT time_mandante
FROM partidas natual join copasdoMundo
WHERE gols_marcados >= 150
"""

df = pd.read_sql_query(query, conn)
df
```

```python
query = """
SELECT DISTINCT iniciais_time, nome_treinador
FROM partidas as p natural join jogadores as j
WHERE p.ano = 1970 AND p.fase = 'Quartas de Final'
"""

df = pd.read_sql_query(query, conn)
df
```

```python
query = """
SELECT DISTINCT cidade, estadio, publico
FROM partidas natural join local
WHERE ano = 2006 AND (gols_mandante >= 3 OR gols_visitante >= 3)
"""

df = pd.read_sql_query(query, conn)
df
```

```python
query = """
SELECT DISTINCT juiz, data_hora, publico
FROM partidas natural join arbitragem
WHERE ano = 2010
ORDER BY publico DESC
LIMIT 5
"""

df = pd.read_sql_query(query, conn)
df
```

```python
query = """
SELECT DISTINCT juiz, assistente_um, assistente_dois
FROM partidas natural join arbitragem
WHERE ano = 2014 AND (fase = 'Final' OR fase = 'Semifinal')
"""

df = pd.read_sql_query(query, conn)
df
```
